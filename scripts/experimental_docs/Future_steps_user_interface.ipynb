#### 4. Run the model (e.g. SIR model) in stochastic and compare the stochastic results to the deterministic results for a set of parameters.

<div style="font-size:14px; font-family: Consolas, 'Fira Code', monospace;">

```python
#Run the SIR model in deterministic way and save the results.
deterministic_results = mw_SIR.run(init_population, timeframe, parameters, stochastic = False)

#Reset the model to its initial state.
mw_SIR.reset()

#Run the SIR model in stochastic way and save the results.
stochastic_results = mw_SIR.run(init_population, timeframe, parameters, stochastic = True)

#Plot the S, I, R data of deterministic and stochastic results along with time.
plt(deterministic_results, stochastic_results)
```
