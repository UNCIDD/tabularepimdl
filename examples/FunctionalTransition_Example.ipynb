```python
import numpy as np
import pandas as pd
from tabularepimdl.FunctionalTransition import FunctionalTransition
```

```python
#population
input_pop = pd.DataFrame({
    "InfState": ["S", "I"],
    "N": [80, 20]
})

# context setup
lockdown_context = {
    "base_beta": 0.5,
    "suppression": 0.3,
    "lockdown_start": 15
}
```

```python
#time-based policy shift based on global time (e.g., lockdown starts at time t=15)
def time_sensitive_infection(state: pd.DataFrame, t: float, context: dict) -> float:
    beta = context["base_beta"]
    suppression = context["suppression"]
        
    if t >= context["lockdown_start"]:
        beta = beta * suppression

    I = state[state["InfState"] == "I"]["N"].sum()
    N = state["N"].sum()
    return beta * I / N
```

```python
time_sensitive_result = []
column_names = ['Time', 'Transmission_Rate', 'Infected_Individuals']
```

```python
timeshift_rule = FunctionalTransition(column='InfState', from_st='S', to_st='I', rate=time_sensitive_infection)
for i in range(0, 4, 1): 
    time_sensitive_result.append(timeshift_rule.get_deltas(current_state=input_pop, dt=7, context=lockdown_context))

time_sensitive_df = pd.DataFrame(time_sensitive_result, columns=column_names)
time_sensitive_df
```
